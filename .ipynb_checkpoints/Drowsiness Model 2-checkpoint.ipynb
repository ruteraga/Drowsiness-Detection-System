```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import cv2
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from keras.utils import to_categorical
import random
import pandas as pd
import dlib
import face_recognition
from scipy.spatial import distance 
import math
from imutils import face_utils
from sklearn.metrics import accuracy_score
import joblib
```

```python
# Train dataset
train_dataset_dir = r"C:\Users\grgabin\Desktop\Projects\Thesis\Dataset\dataset_new\train"
# Test dataset
test_dataset_dir = r"C:\Users\grgabin\Desktop\Projects\Thesis\Dataset\dataset_new\test"
```

```python
# Using a pre-trained model to extract face landmarks
p = "shape_predictor_68_face_landmarks.dat"
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(p)
```

```python
# Plotting the 68 landmarks found on the image
img_path=r'C:\Users\grgabin\Desktop\Projects\Thesis\Dataset\dataset_new\train\yawn\31.jpg'
image= cv2.imread(img_path)
image_rgb=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
rects=detector(image, 0)
plt.figure(figsize=(8,6))
plt.imshow(image_rgb)
for (i, rect) in enumerate(rects):
    shape=predictor(gray, rect)
    shape = predictor(gray, rect)
    shape = face_utils.shape_to_np(shape)
    for(x, y) in shape:
        plt.scatter(x,y, color='green', s=5)
```

```python
#Defining function to get mouth_aspect_ratio: mar
def mouth_aspect_ratio(mouth):
    A = distance.euclidean(mouth[14], mouth[18])
    C = distance.euclidean(mouth[12], mouth[16])
    mar = (A ) / (C)
    return mar
```

```python
classes=[]
data=[]
features=[]
labels=[]
image_size=(224,224)

# Load Dataset and Normalize
for path, subdirs, files in os.walk(train_dataset_dir):
    num_files = len(files)
    num_to_select = num_files
    selected_files = random.sample(files, num_to_select)
        
    for name in tqdm(selected_files):
        img_path = os.path.join(path, name)
        img= cv2.imread(img_path)
        img= cv2.resize(img, (224,224))
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        rects=detector(img, 0)
        for (i, rect) in enumerate(rects):
            shape=predictor(gray, rect)
            shape = predictor(gray, rect)
            shape = face_utils.shape_to_np(shape)
            data.append(shape)
            path_parts = path.split('/')
            label = path_parts[-1]
            if label not in classes:
                classes.append(label)
            index = classes.index(label)
            labels.append(index)
```

```python
features = pd.DataFrame(columns=["MAR"])
for d in data:
    eye=d[36:68]
    mar=mouth_aspect_ratio(eye)
    #mouth_eye=mouth_over_eye(eye)
    #new_row = ({"EAR":ear,"MAR": mar,"Circularity": cir,"MOE": mouth_eye})
    new_row = ({"MAR":mar})
    features.loc[len(features)] = new_row
```

```python
features.fillna(0, inplace=True)
X_train = np.asarray(features)
y_train = np.asarray(labels)
```

```python
y_train=to_categorical(y_train)
```

```python
# Repeating the same process for the test dataset

classes2=[]
data2=[]
features2=[]
labels2=[]
image_size=(224,224)

# Load Dataset and Normalize
for path, subdirs, files in os.walk(test_dataset_dir):
    num_files = len(files)
    num_to_select = num_files
    selected_files = random.sample(files, num_to_select)
        
    for name in tqdm(selected_files):
        img_path = os.path.join(path, name)
        img= cv2.imread(img_path)
        img= cv2.resize(img, (224,224))
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        rects=detector(img, 0)
        for (i, rect) in enumerate(rects):
            shape=predictor(gray, rect)
            shape = predictor(gray, rect)
            shape = face_utils.shape_to_np(shape)
            data2.append(shape)
            path_parts = path.split('/')
            label2 = path_parts[-1]
            if label2 not in classes2:
                classes2.append(label2)
            index = classes2.index(label2)
            labels2.append(index)
```

```python
features2 = pd.DataFrame(columns=["MAR"])
for d in data2:
    eye=d[36:68]
    mar=mouth_aspect_ratio(eye)
    new_row = ({"MAR": mar})
    features2.loc[len(features2)] = new_row
```

```python
features2.fillna(0, inplace=True)
X_test = np.asarray(features2)
y_test = np.asarray(labels2)
```

```python
y_test=to_categorical(y_test)
```

```python
best_accuracy=0
best_k=0
for k in range(1,100):
    knn_classifier=KNeighborsClassifier(n_neighbors=k)
    
    knn_classifier.fit(X_train, y_train)
    y_pred=knn_classifier.predict(X_test)
    accuracy=accuracy_score(y_test, y_pred)
    
    if accuracy > best_accuracy:
        best_accuracy=accuracy
        best_k=k
```

```python
print(best_accuracy)
print(best_k)
```

```python
model=KNeighborsClassifier(n_neighbors=best_k)
model.fit(X_train, y_train)
```

```python
joblib.dump(model, 'knn_model2.pkl')
```
